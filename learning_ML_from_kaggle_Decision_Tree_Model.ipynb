```python




import pandas as pd

data_path = '/content/winequality.csv'

wine_data = pd.read_csv(data_path)

# dataframe of wine dataset

wine_data.describe()
```

```python
# showing the columns of data
wine_data.columns
```

```python
# printing the missing values from data
wine_data = wine_data.dropna(axis=0)
```

```python
wine_data
```

```python
wine_data.describe()
```

```python


# Selecting the prediction target

y =  wine_data.quality
```

```python
# choosing the features 

wine_quality_data_features = [ 'density' , 'pH' , 'alcohol', 'volatile acidity','fixed acidity' ]
```

```python
# for convienace nameing features as X


X  = wine_data[wine_quality_data_features]
```

```python
X.describe()
```

```python
X.head()   # this method shows the top five rows of the data
```

```python
# Building Model using Scikit-Learn
# Scikit-learn is easily the most popular library for modeling the types of typically stores in the DataFrames.

```

The Steps required to build the ML model are:
1. **Define** : What type of model will it be? A decision tree? or other? also some parameters are specified too.
2. **fit** : Capturing the data from provided data. This is the heart of modeling.
3. **Predict**: Just What it sounds like
4. **Evaluate** : Determine how accurate the model's predictions are.

```python
import sklearn
from sklearn.tree import DecisionTreeRegressor

# 1. Define Model


wine_quality_model = DecisionTreeRegressor(random_state=1)


#2. Fit Model

wine_quality_model.fit(X,y)



```
